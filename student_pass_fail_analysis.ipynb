```python
import pandas as pd 
import numpy as np  
from sklearn.metrics import accuracy_score
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("student_exam_performance_dataset.csv")
df=pd.DataFrame(data[['study_hours_per_day','attendance_rate','participation_score','parental_education','pass_fail']])

pass_fail=[]
par_edu=[]
for i in df['pass_fail']:
    if i=='Pass':
        pass_fail.append(1)
    else:
        pass_fail.append(0)

for i in df['parental_education']:
    if i=='High School':
        par_edu.append(0)
    elif i=='Bachelor':
        par_edu.append(1)
    elif i=='Master':
        par_edu.append(2)
    else:
        par_edu.append(3)
df['par_edu']=par_edu
x_train=df.drop(columns=['pass_fail','parental_education'])[:round(len(df['participation_score'])/100)*80]
x_test=df.drop(columns=['pass_fail','parental_education'])[round(len(df['participation_score'])/100)*80:]
y_train=pass_fail[:round(len(df['participation_score'])/100)*80]
y_test=pass_fail[round(len(df['participation_score'])/100)*80:]

model=linear_model.LogisticRegression()
model.fit(np.array(x_train).reshape(8000,4),np.array(y_train))
y_pred=model.predict(np.array(x_test).reshape(2000,4))
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)

for i in model.predict_proba(x_test):
    print(i)
```
